```python
import numpy as np
from skimage import data
import matplotlib.pyplot as plt
import random
import cv2
```

```python
image = data.camera()
for i in range(image.shape[0]):
    for j in range(image.shape[1]):
        image[i,j] += random.randint(-20,20)
```

```python
plt.imshow(image, cmap = plt.cm.gray)
```

```python
mm = np.median(image)
print(mm)
```

```python
binary = image > mm
```

```python
plt.imshow(binary, cmap = plt.cm.gray)
```

```python
from skimage import feature

edges1 = feature.canny(image, sigma=2, low_threshold=1, high_threshold=145)
```

```python
plt.imshow(edges1, cmap = plt.cm.gray)
```

```python
# Метод Отцу
high_thresh, thresh_im = cv2.threshold(image, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
lowThresh = 0.5*high_thresh
```

```python
from skimage import feature

edges1 = feature.canny(image, sigma=2, low_threshold=lowThresh, high_threshold=high_thresh)
```

```python
plt.imshow(edges1, cmap = plt.cm.gray)
```
